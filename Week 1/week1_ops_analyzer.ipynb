This notebook automates the tracking of operational efficiency across different business sites. 
Instead of manually calculating performance metrics, this script processes operational outputs, 
calculates key performance indicators (KPIs), and flags areas that require immediate management attention.

We will achieve this through three steps:
1. Creating a mathematical tool to calculate efficiency.
2. Building a classifier to automatically rank performance.
3. Simulating a dataset of 3 distinct operational sites (Fuel Depots) to run the analysis.

```python
def calculate_efficiency(actual_output, target_output):
    """
    Calculates the efficiency percentage of a given site.
    Formula: (Actual / Target) * 100
    """
    # Prevent division by zero errors just in case a site was offline
    if target_output == 0:
        return 0.0
    
    efficiency = (actual_output / target_output) * 100
    return round(efficiency, 2)

def get_operational_status(efficiency):
    """
    Classifies the health of the operation based on its efficiency percentage.
    """
    if efficiency < 70:
        return "Critical"
    elif 70 <= efficiency < 90:
        return "Warning"
    else:
        return "Normal"
```

### Modeling Our Operations Data
To analyze our operations, we need a structured way to hold data for multiple locations. 
We use a **List of Dictionaries**. Think of the list as our master filing cabinet, and each 
dictionary inside it as a specific site's file folder containing its name, actual fuel output, 
and target fuel output (measured in thousands of liters).

```python
# 3 Fuel Depot Sites Dataset
fuel_depots = [
    {
        "name": "Alpha Depot (Coastal)",
        "actual_output": 1450,
        "target_output": 1500
    },
    {
        "name": "Bravo Depot (Inland HUB)",
        "actual_output": 920,
        "target_output": 1200
    },
    {
        "name": "Charlie Depot (Northern Rail)",
        "actual_output": 450,
        "target_output": 800
    }
]

print("==================================================")
print("       WEEKLY OPERATIONAL PERFORMANCE REPORT      ")
print("==================================================\n")

# Loop through each site to calculate metrics and print the formatted report
for depot in fuel_depots:
    # Extract values
    name = depot["name"]
    actual = depot["actual_output"]
    target = depot["target_output"]
    
    # Run our functions
    eff = calculate_efficiency(actual, target)
    status = get_operational_status(eff)
    
    # Print a clean, formatted report
    print(f"Site Name:   {name}")
    print(f"Performance: {actual:,} / {target:,} Liters")
    print(f"Efficiency:  {eff}%")
    print(f"Status:      [{status}]")
    print("-" * 50)
```
